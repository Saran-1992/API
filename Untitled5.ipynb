```python
import torch
import torch.nn as nn
from torch.nn import functional as F

# --- Step 1: Hyperparameters Configuration ---
# Intha constants thaan namma miniature brain-oda dimensional limits-ah mudivu pannum.
BATCH_SIZE = 4       # Sequential optimization loops-kulla parallel-ah poga pora batch count
BLOCK_SIZE = 8       # Context Length - Oru thadava model-ala maximum gavanikka mudiyaadha context tokens length
MAX_ITERS = 1000     # Training epochs optimization loop lines count
LEARNING_RATE = 1e-3 # Brain cell weights mathurathukkana engine speed control step size
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu' # GPU irundha automatic-ah memory vector graphics computing-ku maaridum

print(f"🚀 MicroGPT Core Pipeline Engine Initialized on: {DEVICE}")

# --- Step 2: Virtual Matrix Operations Test ---
# Real text data-va thallுறதுக்கு முன்னாடி, PyTorch Tensors dynamic math vectors-ah epdi handle பண்ணுதுனு செக் பண்ணுவோம்.
torch.manual_seed(1337) # Random numbers eppavumae uniform-ah generator aagurathukkana seed parameter

# Create a dummy batch of inputs (representing tokens strings index IDs)
# Shape: (BATCH_SIZE, BLOCK_SIZE) -> (4, 8)
X_dummy = torch.randint(0, 100, (BATCH_SIZE, BLOCK_SIZE)).to(DEVICE)

print("\n📊 Initial Dummy Input Matrix Layout (X):")
print(X_dummy)
print(f"Matrix Structural Dimensions: {X_dummy.shape}")
```

```python
# --- Step 1.2: Raw Text Data & Character-Level Tokenizer ---

# Sample Text Dataset (உங்க ப்ராஜெக்ட் மற்றும் காலேஜ் டேட்டா வச்சே ட்ரெய்ன் பண்ணுவோம்!)
raw_text = """
Saran Babu S is a third-year AML student at Muthayammal Engineering College.
He is building an advanced Generative AI model named MicroGPT.
AI and Machine Learning engineering is the core architecture of future technology.
Muthayammal Engineering College is located in Namakkal.
"""

# Text-ல இருக்குற மொத்த தனித்துவமான எழுத்துக்களை (Unique Characters) மட்டும் தனியா பிரிக்கிறோம்
chars = sorted(list(set(raw_text)))
vocab_size = len(chars)
print(f"🔤 Total Unique Characters (Vocab Size): {vocab_size}")
print(f"🎨 All Characters Found: {''.join(chars)}\n")

# --- Encoding & Decoding Mapping (The Tokenizer Engine) ---
# எழுத்துக்களை நம்பராகவும், நம்பரை எழுத்தாகவும் மாற்றும் அகராதி (Dictionaries)
char_to_int = { ch:i for i,ch in enumerate(chars) }
int_to_char = { i:ch for i,ch in enumerate(chars) }

# Encoder: Text strings-ஐ வாங்கி Number IDs கொண்ட List-ஆக மாற்றும்
encode = lambda s: [char_to_int[c] for c in s]

# Decoder: Number IDs கொண்ட List-ஐ வாங்கி திரும்பவும் மனித வார்த்தைகளாக மாற்றும்
decode = lambda l: ''.join([int_to_char[i] for i in l])

# Test Your Tokenizer!
test_string = "Saran Babu"
encoded_vector = encode(test_string)

print(f"📝 Original Text: '{test_string}'")
print(f"🔢 Encoded Matrix Vector: {encoded_vector}")
print(f"🔄 Decoded Back to Text: '{decode(encoded_vector)}'")
```

```python
# --- Step 1.3: Data Tensor Splitting & Batching Ingestion ---

# மொத்த raw text-ஐயும் எண்களா மாத்தி PyTorch Tensor-ஆ மாத்துறோம்
data = torch.tensor(encode(raw_text), dtype=torch.long)

# 90% Data-வை ட்ரெய்னிங்கிற்கும், 10% Data-வை வேலிடேஷனுக்கும் பிரிக்கிறோம்
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]

print(f"📦 Total Tokens in Dataset: {len(data)}")
print(f"📉 Train Data Size: {len(train_data)} tokens")
print(f"📊 Validation Data Size: {len(val_data)} tokens\n")

# --- Dynamic Data Batching Function ---
# மாடல் மூளைக்குள்ள $4 \times 8$ கிரிட் பியாண்ட் இன்புட்களை லோட் செய்யும் இன்ஜின்
def get_batch(split):
    # Train அல்லது Val split செலக்ட் செய்கிறோம்
    data_set = train_data if split == 'train' else val_data

    # ரேண்டமா 4 இடங்களை (Batch Size) செலக்ட் பண்றோம், அங்க இருந்து 8 வார்த்தைகளை (Block Size) எடுக்க
    ix = torch.randint(len(data_set) - BLOCK_SIZE, (BATCH_SIZE,))

    # X (Input): தற்போதைய 8 டோக்கன்கள்
    x = torch.stack([data_set[i:i+BLOCK_SIZE] for i in ix])

    # Y (Target): அடுத்த ஒரு வார்த்தை தள்ளி இருக்கும் 8 டோக்கன்கள் (Target for prediction)
    y = torch.stack([data_set[i+1:i+BLOCK_SIZE+1] for i in ix])

    # GPU-க்கு டேட்டாவை புஷ் பண்றோம்
    return x.to(DEVICE), y.to(DEVICE)

# ஒரு சாம்பிள் பேட்ச் எடுத்து அவுட்புட்டைப் பார்ப்போம்!
xb, yb = get_batch('train')
print("🔥 Input Context Batch (X):")
print(xb)
print("\n🎯 Target Next-Token Batch (Y):")
print(yb)
```

```python
# --- Step 1.4: Building a Miniature Bigram Language Model ---

class MicroGPTModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        # Token Embedding Table: ஒவ்வொரு டோக்கன் ஐடிக்கும் நேரா அதுக்கான அர்த்த வெக்டர்களை மேப் பண்ணும்
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)

    def forward(self, idx, targets=None):
        # idx மற்றும் targets ரெண்டுமே (B, T) வடிவத்துல இருக்குற மேட்ரிக்ஸ்
        logits = self.token_embedding_table(idx) # வடிவமைப்பு: (Batch, Time, Channels) -> (4, 8, 37)

        if targets is None:
            loss = None
        else:
            # PyTorch-ஓட CrossEntropyLoss கணக்குக்காக மேட்ரிக்ஸ் பரிமாணங்களை மாத்துறோம்
            B, T, C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            # Loss Calculation: மாடல் எந்த அளவுக்கு தப்பா பிரிடிக்ட் பண்ணுதுன்ற கணக்கு
            loss = F.cross_entropy(logits, targets)

        return logits, loss

    def generate(self, idx, max_new_tokens):
        # idx என்பது தற்போதைய இன்புட் (B, T)
        for _ in range(max_new_tokens):
            # பிரிடிக்ஷன்ஸ் எடுக்கிறோம்
            logits, loss = self.forward(idx)
            # கடைசி டைம் ஸ்டெப்ல இருக்குற லாஜிக்ஸை மட்டும் போக்கஸ் பண்றோம்
            logits = logits[:, -1, :] # வடிவமைப்பு (B, C)
            # ப்ராபபிலிட்டி (Probability) கணக்கிடுறோம்
            probs = F.softmax(logits, dim=-1) # (B, C)
            # அடுத்த டோக்கனை சாம்பிள் பண்றோம்
            idx_next = torch.multinomial(probs, num_samples=1) # (B, 1)
            # வந்த டோக்கனை பழைய இன்புட் கூட சேர்த்துட்டே போறோம்
            idx = torch.cat((idx, idx_next), dim=1) # (B, T+1)
        return idx

# மாடலை உருவாக்கி GPU-க்கு அனுப்புறோம்!
model = MicroGPTModel(vocab_size).to(DEVICE)
logits, loss = model(xb, yb)
print(f"📊 Raw Prediction Logits Shape: {logits.shape}")
print(f"📉 Initial Training Loss: {loss.item():.4f}\n")

# ட்ரெய்னிங் இல்லாம சும்மா மாடலை பேச வச்சு பார்ப்போம் (Blank Generation)
context = torch.zeros((1, 1), dtype=torch.long, device=DEVICE) # 0 என்பது ஸ்பேஸ் டோக்கன்
generated_output = decode(model.generate(context, max_new_tokens=100)[0].tolist())

print("🎲 Untrained Model (Blabbering Output):")
print("-" * 40)
print(generated_output)
print("-" * 40)
```

```python
# --- Step 1.5: The Optimization and Training Loop ---

# PyTorch Optimizer: மாடலோட தப்புகளைத் திருத்தி எடைகளை (Weights) மாத்தும் இன்ஜின்
optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)

print("🏋️‍♂️ MicroGPT Training Loop Started...\n")

for steps in range(MAX_ITERS):
    # ட்ரெய்னிங் பேட்ச் டேட்டாவை எடுக்கிறோம்
    xb, yb = get_batch('train')

    # Forward Pass: மாடலை பிரிடிக்ட் பண்ண வச்சு லாஸைக் கணக்கிடுறோம்
    logits, loss = model(xb, yb)

    # Backward Pass: பழசை ரீசெட் பண்ணிட்டு, கிரேடியன்ட் கணக்கு பண்றோம் (Backpropagation)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()

    # Weights Update: மாடல் மூளையோட நியூரான்களை கொஞ்சம் அட்ஜஸ்ட் பண்றோம்
    optimizer.step()

    # ஒவ்வொரு 200 ஸ்டெப்புக்கும் லாஸ் எவ்வளவு குறையுதுன்னு லைவா பார்ப்போம்
    if steps % 200 == 0:
        print(f"🔄 Step {steps:4d} | Current Training Loss: {loss.item():.4f}")

print("\n🏁 Training Completed successfully!")
print(f"📉 Final Training Loss: {loss.item():.4f}\n")

# --- Test Trained Model (The Magic Moment!) ---
print("🌟 Trained Model Output (Smart Generation):")
print("-" * 40)
context = torch.zeros((1, 1), dtype=torch.long, device=DEVICE)
trained_output = decode(model.generate(context, max_new_tokens=150)[0].tolist())
print(trained_output)
print("-" * 40)
```

```python
# --- Step 1.6: Upgrading to Self-Attention Matrix Mechanism ---

class ScaledDotProductAttention(nn.Module):
    def __init__(self, embed_size):
        super().__init__()
        # Query, Key, Value மேட்ரிக்ஸ் எடைகளை உருவாக்குறோம்
        self.query = nn.Linear(embed_size, embed_size, bias=False)
        self.key = nn.Linear(embed_size, embed_size, bias=False)
        self.value = nn.Linear(embed_size, embed_size, bias=False)
        # Future Tokens-ஐ மாடல் பார்க்காமல் மறைக்கும் Mask (Tril)
        self.register_buffer('tril', torch.tril(torch.ones(BLOCK_SIZE, BLOCK_SIZE)))

    def forward(self, x):
        B, T, C = x.shape
        q = self.query(x) # (B, T, C)
        k = self.key(x)   # (B, T, C)
        v = self.value(x) # (B, T, C)

        # Attention Scores கணக்கிடுதல்: வார்த்தைகளுக்கு இடையே உள்ள தொடர்பை எடையிடுதல்
        # Formula: (Q * K^T) / sqrt(d_k)
        wei = q @ k.transpose(-2, -1) * (C ** -0.5) # (B, T, T)

        # Decoder Masking: எதிர்கால வார்த்தைகளை மறைக்கிறோம் (Autoregressive Property)
        wei = wei.masked_fill(self.tril[:T, :T] == 0, float('-inf'))
        wei = F.softmax(wei, dim=-1) # (B, T, T)

        # வெயிட்டுகளை வேல்யூ கூட பெருக்கி ஃபைனல் அவுட்புட் எடுக்கிறோம்
        out = wei @ v # (B, T, C)
        return out

print("💎 Scaled Dot-Product Self-Attention Layer Compiled Ready!")
```

```python
# --- Step 1.7: Upgraded MicroGPT Model with Attention Layer ---

class MicroGPTWithAttention(nn.Module):
    def __init__(self, vocab_size, embed_size=32):
        super().__init__()
        # 1. Token Embedding: டோக்கன் ஐடிகளை எம்பெடிங் வெக்டர்களா மாத்துறோம்
        self.token_embedding_table = nn.Embedding(vocab_size, embed_size)

        # 2. Position Embedding: டிரான்ஸ்ஃபார்மருக்கு வார்த்தைகளோட வரிசை/இடம் தெரியாது, அதனால பொசிஷன் வெக்டரையும் சேர்க்கிறோம்
        self.position_embedding_table = nn.Embedding(BLOCK_SIZE, embed_size)

        # 3. Attention Head: நாம இப்போ கம்பைல் பண்ணின அந்த அட்வான்ஸ்டு லேயர்!
        self.attention_head = ScaledDotProductAttention(embed_size)

        # 4. Language Model Head: அவுட்புட்டை திரும்பவும் Vocab Size-க்கு மாத்துற லீனியர் லேயர்
        self.lm_head = nn.Linear(embed_size, vocab_size)

    def forward(self, idx, targets=None):
        B, T = idx.shape

        # டோக்கன் மற்றும் பொசிஷன் எம்பெடிங்களை கூட்டி ஒன்னாக்குறோம்
        tok_emb = self.token_embedding_table(idx) # (B, T, embed_size)
        pos_emb = self.position_embedding_table(torch.arange(T, device=DEVICE)) # (T, embed_size)
        x = tok_emb + pos_emb # (B, T, embed_size)

        # Self-Attention லேயர் வழியா டேட்டாவை பாஸ் பண்றோம் (இப்போ மாடல் முழு கான்டெக்ஸ்ட்டையும் கவனிக்கும்!)
        x = self.attention_head(x) # (B, T, embed_size)

        # லாஜிட்களை கணக்கிடுறோம்
        logits = self.lm_head(x) # (B, T, vocab_size)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            loss = F.cross_entropy(logits, targets)

        return logits, loss

    def generate(self, idx, max_new_tokens):
        for _ in range(max_new_tokens):
            # கான்டெக்ஸ்ட் சைஸ் BLOCK_SIZE-ஐ விட தாண்டாம இருக்க கிராப் பண்றோம்
            idx_cond = idx[:, -BLOCK_SIZE:]
            logits, loss = self.forward(idx_cond)
            logits = logits[:, -1, :] # (B, C)
            probs = F.softmax(logits, dim=-1) # (B, C)
            idx_next = torch.multinomial(probs, num_samples=1) # (B, 1)
            idx = torch.cat((idx, idx_next), dim=1) # (B, T+1)
        return idx

# புதிய மாடலை உருவாக்கி GPU-க்கு அனுப்புறோம்!
model_v2 = MicroGPTWithAttention(vocab_size).to(DEVICE)
print("🚀 MicroGPT_V2 with Self-Attention Engine Armed and Ready!")
```

```python
# --- Step 1.8: Training Loop for MicroGPT_V2 ---
optimizer_v2 = torch.optim.AdamW(model_v2.parameters(), lr=LEARNING_RATE)

print("🏋️‍♂️ Training MicroGPT_V2 with Attention Loop Started...\n")

for steps in range(MAX_ITERS):
    xb, yb = get_batch('train')
    logits, loss = model_v2(xb, yb)

    optimizer_v2.zero_grad(set_to_none=True)
    loss.backward()
    optimizer_v2.step()

    if steps % 200 == 0:
        print(f"🔄 Step {steps:4d} | Current Training Loss: {loss.item():.4f}")

print("\n🏁 Training Completed!")
print(f"📉 Final Training Loss: {loss.item():.4f}\n")

# --- Test the Smart Upgraded Model! ---
print("🌟 Upgraded Model Output (Smart Attention Generation):")
print("-" * 40)
context = torch.zeros((1, 1), dtype=torch.long, device=DEVICE)
trained_output_v2 = decode(model_v2.generate(context, max_new_tokens=150)[0].tolist())
print(trained_output_v2)
print("-" * 40)
```
